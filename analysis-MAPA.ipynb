```python
# Donkey Slaughter Analysis — MAPA Data

This notebook analyzes data on donkey slaughter in Brazil, based on records from the Ministry of Agriculture and Livestock (MAPA).

## Goal

Identify which slaughterhouses registered the highest volume of donkey slaughter in 2025, also calculating the monthly average for each facility.

## Notebook structure

1. Data loading and initial exploration
2. Filtering for the year 2025
3. Aggregation by facility and monthly average calculation


import pandas as pd

df = pd.read_csv('abate_asinino.csv')
df
```

```python
#Need to select only 2025 records

df_2025 = df[df['Abate'].str.contains('-25')]
```

```python
print (df_2025)
```

```python

df.columns
```

```python


#Group the data by facility and sum the total slaughter volume for 2025
#calculate the monthly average 

media_2025 = df_2025.groupby(['SIF', 'Estabelecimento', 'UF', 'Município'])['Qtd Abate'].sum().reset_index()
media_2025['Média mensal'] = media_2025['Qtd Abate'] / 12
media_2025 = media_2025.sort_values('Qtd Abate', ascending=False)
```

```python
print (media_2025)
```

```python

##The table shows the facilities ranked by total slaughter volume in 2025.


media_2025 = media_2025.sort_values('Qtd Abate', ascending=False).reset_index(drop=True)
media_2025
```
